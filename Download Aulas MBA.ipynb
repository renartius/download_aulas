# get files and save them

```python
def getMaterial(material, classId, disciplineName, folder=None):
    req = s.get(material['Url'])
    title = material['DownloadTitle']    
    time.sleep(1)    
    if folder is not None:
        filename = f'{remove_special_chars(disciplineName)}/{remove_special_chars(folder)}/{remove_special_chars(title)}'
    else:
        filename = f'{remove_special_chars(disciplineName)}/{remove_special_chars(title)}'
    print(filename)    
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    with open(filename, 'wb') as f:
        f.write(req.content)


def remove_special_chars(filename):
    for char in ['\\','/',':','*','?','|','"','<','>']:
        filename = filename.replace(char, '')
    return filename.rstrip()


for obj in r.json():
    if obj['title']=="Aula":
        if obj["hasMaterial"]:
            print(obj['extra']['disciplineName'])
            print(obj['extra']['classId'])
            classId = obj['extra']['classId']
            disciplineName = obj['extra']['disciplineName']
            rml = s.get(f'{urlMaterialsList}{classId}')
            #print(rml.json())            
            #break
            time.sleep(1)
            for material in rml.json():
                mat_url = material['Url']
                folder = None
                if mat_url is not None:
                    getMaterial(material, classId, disciplineName, folder)                                    
                hasChildren = material['HasChildren']
                if hasChildren:
                    folder = material['Title']                                 
                    for child in material['Children']:
                        getMaterial(child, classId, disciplineName, folder)                         
```
